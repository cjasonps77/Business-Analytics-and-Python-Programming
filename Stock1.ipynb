1. Download Data
Please fetch the data of Amazon, Microsoft and Google from 1st Jan 2010 to 30th Dec 2020. 
Show a table of “Close price” of all stocks in one table.

```python
import pandas as pd
import numpy as np
import yfinance as yf

tickers = ["AMZN", "MSFT", "GOOGL"]
data = yf.download(tickers, start="2010-01-01", end="2020-12-30")
data = data[["Close"]]
data
```

2. Calculate daily returns of each stock.
Calculate daily returns and delete the row of ‘NaN’ data.
Show a table of daily returns of all stocks
Create portfolios of each two-stock with the same proportion for each stock, i.e. (50% Amazon, 50% Microsoft), (50% Amazon, 50% Google), (50% Microsoft, 50% Google). 

```python
dailyreturn = data.pct_change()
dailyreturn = dailyreturn.dropna()
dailyreturn
```

```python
proportion = [0.5,0.5,0.5]
combinedreturns = (proportion * dailyreturn) 
combinedreturns

stock1 = combinedreturns[[('Close', 'AMZN'),('Close', 'MSFT')]].sum(axis=1)
stock2 = combinedreturns[[('Close', 'AMZN'),('Close', 'GOOGL')]].sum(axis=1)
stock3 = combinedreturns[[('Close', 'MSFT'),('Close', 'GOOGL')]].sum(axis=1)

combinedstockreturn = pd.concat([stock1, stock2, stock3],axis=1)
combinedstockreturn.rename(columns={0:'A_M',1:'A_G',2:'M_G'},inplace=True)
combinedstockreturn
```

3. Calculate CV
Calculate CV of three portfolios above. 
Which portfolio is the most effective one based on CV?

```python
CV = combinedstockreturn.std()/combinedstockreturn.mean()
print('CV of each portfolio: \n\n',round(CV, 3))
print('The most effective portofolio based on CV is A_M which is the combination of AMAZON and MICROSOFT with the value of:', round(CV.min(), 3))
```

4. Calculate Sharpe Ratio
Calculate Sharpe ratios of the portfolios.
Assuming the risk-free returns is 1% per year (360 days). Which portfolio should investors choose based on Sharpe ratios?

```python
riskfree = 0.01/360
sharpe = (combinedstockreturn.mean() - riskfree) / combinedstockreturn.std()
print('Sharpe ratio of each portfolio:\n',round(sharpe, 3))
print('The portfolio that investors should choose based on Sharpe ratios is A_M which is the combination of AMAZON and MICROSOFT with the value of:', round(sharpe.max(), 3),". Because it has the highest value.")  
```
